### Data Cleaning with Jeopardy!

First we load the data and inspect the head to make sure it is loaded properly. We'll also use `.info` to see more details about the columns.

```python
import pandas as pd

# Loading the data
jeopardy_data = pd.read_csv("jeopardy.csv")
print(jeopardy_data.info())
jeopardy_data.head()
```

We know that the `Dtype` of `"Value"` is `object`, and we can see that `int64` is an available `Dtype`. That should give us a good indication that we won't be able to get numerical descriptive statistics about it. But let's confirm that. Let's try to get the average `"Value"`.

```python
jeopardy_data[" Value"].mean()
```

As we expected, we got an error message related to the type of the column: `could not convert string to float`. We're trying to find the `mean()` of a String. Let's try to convert that column to a number. From looking at the head of the data, we can see that these values are starting with a `$` &mdash; that's going to be a problem. Let's take a look at the other possible values for `"Value"` to see if there's anything else we need to watch out for

```python
jeopardy_data[" Value"].unique()
```

Those commas will have to be removed as well. That `"None"` is also interesting. It might be worthwhile to see how many of those exist, and why they exist. (This is a great example of incomplete data that isn't immediately obvious &mdash; the `.info()` function told us that there were no non-null values for the `"Value"` column, but in reality this missing data was represented by the String `"None"` rather than a `null` value. Let's filter our dataset to show only the rows that contain that `"None"`

```python
jeopardy_data[jeopardy_data[" Value"] == "None"]
```

Ah! All of the Final Jeopardy questions have `"None"` as their value. That makes sense. At this point, we can decide what to do with those values. We have a few options &mdash; we could turn the `"None"`s into acutal `null`s and then filter by `null`. We could turn the `"None"`s into `0`s, but that might effect our statistics. For now, let's just start working with a subset of the data that doesn't contain any of the `"None"` values.

```python
no_final_jeopardy = jeopardy_data[jeopardy_data[" Value"] != "None"]
no_final_jeopardy[" Value"].unique()
```

Nice! We're finally ready to transform this column into a number. The plan is to remove the dollar sign, the commas, and then cast that value to an integer (or float). Let's prove to ourselves we can do this on one example, and then we'll figure out how to apply that example to the entire DataFrame

```python
example_string = "$4,238"
example_string = int(example_string.replace("$","").replace(",",""))
print(example_string)
```

```python
def str_to_int(input_str):
    return int(input_str.replace("$","").replace(",",""))

no_final_jeopardy["Int Value"] = no_final_jeopardy[' Value'].apply(str_to_int)
no_final_jeopardy["Int Value"]
```

Great work! We can now see descriptive statistics for this numeric column by using `.describe()`

```python
no_final_jeopardy.describe()
```

Now that we know we can use the `"Int Value"` column for calculations, let's see if we can filter our data by the question's text. Let's see if we can find all questions that have the word `"computer"` in it.

```python
contains_computer = no_final_jeopardy[" Question"].apply(lambda x: "computer" in x)
computer_rows = no_final_jeopardy[contains_computer]
computer_rows
```

Awesome! We did it! Now let's see how valuable those `"computer"` questions are on average. We can also get a sense of the answers that come up most often when we see the word `"computer"`

```python
print(computer_rows["Int Value"].mean())
print(computer_rows[" Answer"].value_counts())
```
